```python
import matplotlib.pyplot as plt
from math import sin,cos,degrees,radians,e
```

```python
plt.rcParams['xtick.bottom'] = False
plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['ytick.left'] = False
plt.rcParams['ytick.labelleft'] = False
plt.rcParams['axes.edgecolor'] = 'white'

plt.rcParams['axes.grid'] = True
```

```python
def sine_graph(degree):
  """
  Draw a graph of the sine function.

  Args:
    int: degree
  Return:
    None: The sine graph is drawn
  """
  #WRITE YOUR CODE HERE
  plt.plot([0,degree],[0,0],linestyle='dashed',color='black',marker='o')
  x = []
  y = []
  for i in range(0,degree+1):
    x.append(i)
    y.append(sin(radians(i)))
    if i%90==0:
      plt.text(i,-0.1,f'{i}Â°')
      plt.text(i,-0.02,'|')
  plt.plot(x,y)
  plt.show()
```

```python
plt.figure(dpi=150)
sine_graph(720)
```

```python
plt.figure(dpi=150)
sine_graph(180)
```

```python
def cosine_graph(degree):
  """
  Draw a graph of the cosine function.

  Args:
    int: degree
  Return:
    None: The sine graph is drawn
  """
  #WRITE YOUR CODE HERE
  plt.plot([0,degree],[0,0],linestyle='dashed',color='black',marker='o')
  x = []
  y = []
  for i in range(0,degree+1):
    x.append(i)
    y.append(cos(radians(i)))
    if i%90==0:
      plt.text(i,-0.1,f'{i}Â°')
      plt.text(i,-0.02,'|')
  plt.plot(x,y)
  plt.show()
```

```python
plt.figure(dpi=150)
cosine_graph(540)
```

```python
def sine_and_cosine_graph(degree):
  """
  Draw a graph of the sine and cosine function.

  Args:
    int: degree
  Return:
    None: The sine graph is drawn
  """
  #WRITE YOUR CODE HERE
  plt.plot([0,degree],[0,0],linestyle='dashed',color='black',marker='o')
  x = []
  y_sin = []
  y_cos = []
  for i in range(0,degree+1):
    x.append(i)
    y_sin.append(sin(radians(i)))
    y_cos.append(cos(radians(i)))
    if i%90==0:
      plt.text(i,-0.1,f'{i}Â°')
      plt.text(i,-0.02,'|')
  plt.plot(x,y_sin)
  plt.plot(x,y_cos)
  plt.show()
```

```python
plt.figure(dpi=150)
sine_and_cosine_graph(540)
```

```python
plt.figure(dpi=150)
sine_and_cosine_graph(360)
```
